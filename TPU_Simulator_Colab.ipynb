## Load Dataset

This project does not use a traditional dataset file. Instead, we use generated matrices and the project scripts. Here we load the simulator code and sample matrices to verify everything works.

```python
from matrix_generator import generate_matrices

# Generate a sample matrix pair to check imports
mat_a, mat_b = generate_matrices(3)
print('Sample Matrix A:')
print(mat_a)
print('\nSample Matrix B:')
print(mat_b)
```

## Data Cleaning and Preprocessing

Because this is a simulator project, the main input is matrix data. We keep preprocessing simple and show how the code handles matrices that are already ready for multiplication.

```python
# In this project we use clean generated matrices, so there is no extra cleaning step.
# If we had a dataset file, we would handle missing values, convert types, and check shapes here.

def preprocess_matrix(mat):
    return np.array(mat, dtype=float)

mat_a_clean = preprocess_matrix(mat_a)
mat_b_clean = preprocess_matrix(mat_b)
print('Matrix A cleaned shape:', mat_a_clean.shape)
print('Matrix B cleaned shape:', mat_b_clean.shape)
```

## Exploratory Data Analysis

We do a quick check of the matrix values and show a simple visualization. This helps us understand the data shape and the values before multiplication.

```python
import seaborn as sns

print('Statistics for Matrix A:')
print('min=', mat_a_clean.min(), 'max=', mat_a_clean.max(), 'mean=', mat_a_clean.mean())

print('\nStatistics for Matrix B:')
print('min=', mat_b_clean.min(), 'max=', mat_b_clean.max(), 'mean=', mat_b_clean.mean())

plt.figure(figsize=(6, 4))
sns.heatmap(mat_a_clean, annot=True, cmap='viridis')
plt.title('Matrix A Heatmap')
plt.show()

plt.figure(figsize=(6, 4))
sns.heatmap(mat_b_clean, annot=True, cmap='viridis')
plt.title('Matrix B Heatmap')
plt.show()
```
